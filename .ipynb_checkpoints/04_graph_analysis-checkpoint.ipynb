```python
import pandas as pd
import networkx as nx
```

```python
# Cargamos el grafo en memoria
df = pd.read_csv('dataset/dataset_edges.csv')
df.head()
```

```python
graph = nx.read_edgelist("dataset/dataset_edges.csv", delimiter=",", data=[("weight", float)])

print("El grafo tiene {} aristas y {} vértices".format(len(graph.edges(data=True)), len(graph.nodes())))
```

```python
print("Arista (usuario1 -> usuario2) = Nº interacciones")
for (s, t, w) in graph.edges(data="weight"):                 
    if w > 30:                                                                                              
        print("Peso de la arista ({} -> {}) = {}".format(s, t, w))
```

```python
counter = 0
for (node, val) in graph.degree():
    if val > 1:
        counter = counter + 1
        print("Nodo %s con grado = %d" %(node, val))

print(counter)
```
