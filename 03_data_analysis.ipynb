```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/dataset_edges.csv')
```

```python
graph = nx.read_edgelist("dataset/dataset_edges.csv", delimiter=",", data=[("weight", float)])

print("Grafo con {} aristas y {} vértices".format(len(graph.edges(data=True)), len(graph.nodes())))
```

```python
print("Arista (usuario1 -> usuario2) = Nº interacciones")
for (s, t, w) in graph.edges(data="weight"):                 # Iterar sobre las aristas recuperando el peso de las mismas
    if w > 45:                                               # si data=False (por defecto), solo devuelve los nodos 
                                                             # origen y destino de la arista
        print("Peso de la arista (%s -> %s) = %f" % (s, t, w))
```

```python
counter = 0
for (node, val) in graph.degree():
    if val > 1:
        counter = counter + 1
#         print("Nodo %s con grado = %d" %(node, val))

print(counter)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(30, 30))
nx.draw_networkx(graph, with_labels=False,
                 ax=ax, 
                 node_size=50,
                 font_size=10, 
                 alpha=.5)

ax.set_axis_off()
```
